```python
import pandas as pd
from pathlib import Path
```

## Read raw data file

```python
filename = "./data/raw/store47-2016-head.csv"
fileextension = Path(filename).suffix

if fileextension == ".csv":
    df = pd.read_csv (filename)
    print ("csv-file imported as dataframe")
else:
    print ("Wrong file extension!")
```

```python
df.head(10)
```

## Split into train and test data set

```python
# Split the data at the date 2017-08-02 (last 14 days of data set)
df_train = df[df['date'] < '2017-08-02']
df_test  = df[df['date'] >= '2017-08-02']
```

## Save train and test data as csv-file

```python
train_filename = "./data/raw/store47-2016-train.csv"
test_filename  = "./data/raw/store47-2016-test.csv"

df_train.to_csv(train_filename, index=False)
df_test.to_csv(test_filename, index=False)
```
